```python
import pandas as pd
```

```python
df = pd.read_csv("results/experiment_18-02-2023 16_08_51.csv",index_col=0)
models = ["leNet", "alexNet", "vgg11", "resNet18"]
datasets = ["MNIST", "FashionMNIST", "CIFAR10", "CIFAR100"]
quantizations = ["float32", "torch.qint8", "torch.float16", "float64"]

df = df.set_index(keys=["Model", "Dataset", "Quantization"])

for feature in ["Prediction Accuracy", "Mean Abs Value of Output", "Inference Time", "Model Size"]:
    new_col = []
    for model in models:
        for dataset in datasets:
            for quantization in quantizations:
                val = df.loc[model, dataset, quantization][feature]*100/df.loc[model, dataset, "float32"][feature]
                new_col.append(val)
    df[f"% {feature} w.r.t float32"] = new_col

        

```

```python
df.to_csv("final_results.csv")
```
